```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer([("one", OneHotEncoder(), [3])], remainder="passthrough")
x=ct.fit_transform(x)
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x.shape," - xshape")
print(y.shape," - yshape")
print(x_train.shape," - x_train shape")
print(x_test.shape," - x_test shape")
print(y_train.shape," - y_train shape")
print(y_test.shape," - y_test shape")
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Model Building
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
# modelname = Sequential()
regression = Sequential()
```

```python
#Adding input, hidden and output layers
regression.add(Dense( units=6 , kernel_initializer= "random_uniform", activation="relu"))
regression.add(Dense( units= 12, kernel_initializer= "random_uniform", activation="relu"))
regression.add(Dense( units= 1, kernel_initializer= "random_uniform", activation="relu"))
```

```python
regression.compile(optimizer = "adam", loss = "mse", metrics=['mse'])
```

```python
regression.fit(x_train,y_train, epochs=10000)
```

```python
ypred = regression.predict(x_test)
```
