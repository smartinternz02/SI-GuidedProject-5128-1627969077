### **Applying ANN Categorical on Bank Dataset**

```python
# 1. Import the libraries
import numpy as np 
import pandas as pd
```

```python
# 2. Reading the datasets 
data = pd.read_csv('bank.csv')
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,16:17].values
```

```python
# Importing libraries to handle the categorical data 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# Handling categorical data 
ct = ColumnTransformer([('one', OneHotEncoder(), [1,2,3,4,6,7,8,10,15])], remainder = 'passthrough')
x = ct.fit_transform(x)

cto = ColumnTransformer([('one', OneHotEncoder(), [0])], remainder = 'passthrough')
y = cto.fit_transform(y)
```

```python
# Importing libraries to handle splitting 
from sklearn.model_selection import train_test_split
```

```python
# 6. Splitting data to train and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
print("x_train shape", "\t", x_train.shape)
print("x_test shape", "\t", x_test.shape)
print("y_train shape", "\t", y_train.shape)
print("y_test shape", "\t", y_test.shape)
print("x shape", "\t", x.shape)
print("y shape", "\t", y.shape)
```

```python
# Importing libraries to handle feature scaling 
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Import deep learning libraries
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense
```

```python
categorical = Sequential()
```

```python
# Input layer
categorical.add(Dense(units = 51, kernel_initializer = 'random_uniform', activation = 'relu'))
# Hidden layer 1  
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
# Hidden layer 2
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
# Output layer 
categorical.add(Dense(units = 2, kernel_initializer = 'random_uniform', activation = 'softmax'))
```

```python
# Compiling model  
categorical.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Training the model 
categorical.fit(x_train, y_train, epochs = 50)
```

```python
# Testing the model with test set 
ypred = categorical.predict(x_test)
```
